```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
acc = pd.read_csv('ACSST1Y2022.S2701-2024-03-17T051056.csv')
# Select only the required columns
columns_to_keep = [
    'Label (Grouping)',
    'Los Angeles County, California!!Total!!Estimate',
    'Los Angeles County, California!!Insured!!Estimate',
    'Los Angeles County, California!!Uninsured!!Estimate'
]

# Filter the DataFrame to include only the selected columns
acc = acc[columns_to_keep]
acc.columns = ['Label', 'total','Insured', 'Uninsured']
```

```python
age = acc.iloc[2:11].reset_index(drop=True)
age['total'] = age['total'].str.replace(',', '').astype(int)
age['Insured'] = age['Insured'].str.replace(',', '').astype(int)
age['Uninsured'] = age['Uninsured'].str.replace(',', '').astype(int)
print(age)
#draw a histogram of the insured and uninsured population
#age.plot(kind='bar', x='Label (Grouping)', y=['Los Angeles County, California!!Insured!!Estimate', 'Los Angeles County, California!!Uninsured!!Estimate'], stacked=True)
age.plot(kind='bar', x='Label', y=['Insured', 'Uninsured'], stacked=True, figsize=(10, 6))
plt.xticks(rotation=45, ha='right')
total_population = age['total'].sum()
for index, row in age.iterrows():
    # Calculate the percentage
    percentage = f"{(row['total'] / total_population) * 100:.1f}%"
    # Get the x position of the label
    x = index-0.2
    # Get the y position of the label, which is the top of the bar (sum of 'Insured' and 'Uninsured')
    y = row['total']+ 20000
    plt.text(x, y, percentage)
plt.xlabel('Age Group')
plt.ylabel('Millions of People')
plt.show()
```
