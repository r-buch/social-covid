```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
merged_data = pd.read_csv('Processed_Final_merged_dataset.csv')
print(merged_data.dtypes)
```

```python
# calculate the sum of the first 36 columns
column_sums = merged_data.iloc[:, 1:37].sum(axis=0)
print(column_sums)
```

```python
#Population distribution by sex
#draw a pie plot of the DP05_0002E and DP05_0003E columns
labels = ['male', 'female']
sizes = [column_sums[1], column_sums[2]]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
#Population distribution by age groups
#draw a histogram of the index 3-16 numbers in the column_sums
column_sums[3:16].plot(kind='bar')
plt.xticks(np.arange(13), ['Under 5 years', '5 to 9 years', '10 to 14 years', '15 to 19 years', '20 to 24 years', '25 to 34 years', '35 to 44 years', '45 to 54 years', '55 to 59 years', '60 to 64 years', '65 to 74 years', '75 to 84 years', '85 years and over'])
plt.xlabel('Age groups')
plt.ylabel('Population')
plt.title('Population distribution by age groups')
plt.show()
```

```python
#Population distribution by race groups
#draw a histogram of the different race groups
selected_column_sums = column_sums.iloc[[17, 18, 19, 24, 32]]
selected_column_sums.plot(kind='bar')
plt.xticks(np.arange(5), ['White', 'Black or African American', 'American Indian and Alaska Native', 'Asian', 'Native Hawaiian and Other Pacific Islander'])
plt.xlabel('Race groups')
plt.ylabel('Population')
plt.title('Population distribution by race groups')
plt.show()
```

```python
#Population distribution by Hispanic or Latino and not Hispanic or Latino
#draw a histogram of the index 19, 20 numbers in the column_sums
column_sums.iloc[[33, 34]].plot(kind='bar')
plt.xticks(np.arange(2), ['Hispanic or Latino', 'Not Hispanic or Latino'])
plt.xlabel('Hispanic or Latino')
plt.ylabel('Population')
plt.title('Population distribution by Hispanic or Latino and not Hispanic or Latino')
plt.show()
```
